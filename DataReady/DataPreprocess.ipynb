```python
temp1=temp.drop(columns=drop_list)
```

```python
temp1.loc[:,'player_positions']=temp1.player_positions.apply(lambda x:str(x).replace(",","-"))
temp1.loc[:,'player_tags']=temp1.player_tags.apply(lambda x:str(x).replace(",","-"))
temp1.loc[:,'player_traits']=temp1.player_traits.apply(lambda x:str(x).replace(",","-"))
```

```python
temp1.loc[:,'league_id']=temp1.league_id.apply(lambda x:str(int(x)) if not math.isnan(x) else float('nan'))
temp1.loc[:,'club_team_id']=temp1.club_team_id.apply(lambda x:str(int(x)) if not math.isnan(x) else float('nan'))
temp1.loc[:,'nation_team_id']=temp1.nation_team_id.apply(lambda x:str(int(x)) if not math.isnan(x) else float('nan'))
temp1.loc[:,'club_contract_valid_until_year']=temp1.club_contract_valid_until_year.apply(lambda x:str(int(x)) if not math.isnan(x) else float('nan'))
temp1.loc[:,'nation_jersey_number']=temp1.nation_jersey_number.apply(lambda x:str(int(x)) if not math.isnan(x) else float('nan'))
```
